```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Load the dataset
df = pd.read_csv('Lightning_NO2_FINAL_Corrected.csv')

# Filter out the placeholder values
df_filtered = df[df['NO2_Value'] != -9999].copy()

# Define 5 zones based on quantiles (quintiles) of the NO2_Value
# pd.qcut divides the data into equal-sized groups
labels = ['Zone 1: Very Low', 'Zone 2: Low', 'Zone 3: Medium', 'Zone 4: High', 'Zone 5: Very High']
df_filtered['NO2_Zone'] = pd.qcut(df_filtered['NO2_Value'], q=5, labels=labels)

# Create a color mapping for the zones
colors = {'Zone 1: Very Low': 'blue', 'Zone 2: Low': 'green', 'Zone 3: Medium': 'yellow', 'Zone 4: High': 'orange', 'Zone 5: Very High': 'red'}
df_filtered['color'] = df_filtered['NO2_Zone'].map(colors)

# Create the geographical plot
plt.figure(figsize=(12, 8))

# We'll create a scatter plot, colored by our new 'color' column
plt.scatter(df_filtered['longitude'], df_filtered['latitude'], c=df_filtered['color'], s=10, alpha=0.7)

# Create a custom legend for the zones
legend_patches = [mpatches.Patch(color=colors[label], label=f'{label}') for label in labels]

# Add the legend to the plot
plt.legend(handles=legend_patches, title='NO2 Concentration Zones', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Classified Geographical Distribution of NO2 Values into 5 Zones')
plt.grid(True)
plt.tight_layout()

# Save and show the plot
plt.savefig('NO2_classified_map.png')
plt.close()
```

```python
import rasterio
from rasterio.warp import reproject, Resampling
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

print("--- Performing Principal Component Analysis (PCA) ---")

try:
    # --- Step 1: Load and Align Your Data Layers ---

    # Open the master raster (NO2) and the one to be resampled (lightning)
    with rasterio.open('NO2_Concentration.tif') as no2_src, \
         rasterio.open('density_map.tif') as density_src:

        # Get the profile and data of the NO2 raster (our template)
        profile = no2_src.profile
        no2_concentration = no2_src.read(1)

        # Create an empty array with the same shape as the NO2 raster
        lightning_resampled = np.empty_like(no2_concentration)

        # Reproject the lightning density map to match the NO2 map's grid
        reproject(
            source=rasterio.band(density_src, 1),
            destination=lightning_resampled,
            src_transform=density_src.transform,
            src_crs=density_src.crs,
            dst_transform=no2_src.transform,
            dst_crs=no2_src.crs,
            resampling=Resampling.bilinear
        )
        print("Successfully aligned raster grids.")

    # Stack the perfectly aligned rasters into a single 3D array
    raster_stack = np.array([lightning_resampled, no2_concentration])

    # Flatten the data for PCA
    n_pixels, n_variables = raster_stack.shape[1] * raster_stack.shape[2], raster_stack.shape[0]
    flat_data = raster_stack.reshape(n_variables, n_pixels).T

    # Remove NoData pixels
    nodata_mask = np.any(flat_data == profile['nodata'], axis=1) | np.isnan(flat_data).any(axis=1)
    valid_data = flat_data[~nodata_mask]

    print(f"Data prepared for PCA with {valid_data.shape[0]} valid pixels and {valid_data.shape[1]} variables.")

    # --- Step 2: Scale the Data and Run PCA ---

    # Standardize the data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(valid_data)

    # Initialize and run PCA
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(scaled_data)

    # --- Step 3: Analyze and Interpret the Results ---

    print("\n--- PCA Results ---")
    explained_variance = pca.explained_variance_ratio_
    print(f"Explained variance by component: {explained_variance}")
    print(f"PC1 captures {explained_variance[0]*100:.2f}% of the total variance.")
    print(f"PC2 captures {explained_variance[1]*100:.2f}% of the total variance.")

    print("\nComponent Loadings:")
    print("                PC1    PC2")
    print(f"Lightning:  {pca.components_[0, 0]:.4f}  {pca.components_[1, 0]:.4f}")
    print(f"NO2:        {pca.components_[0, 1]:.4f}  {pca.components_[1, 1]:.4f}")

    # --- Step 4: Save Principal Components as New GeoTIFFs ---

    # Create empty arrays to hold the full-sized PC maps
    pc1_full = np.full(n_pixels, profile['nodata'], dtype=np.float32)
    pc2_full = np.full(n_pixels, profile['nodata'], dtype=np.float32)

    # Place the calculated PC values back into the non-nodata pixels
    pc1_full[~nodata_mask] = principal_components[:, 0]
    pc2_full[~nodata_mask] = principal_components[:, 1]

    # Reshape the arrays back to 2D raster grids
    pc1_map = pc1_full.reshape(raster_stack.shape[1], raster_stack.shape[2])
    pc2_map = pc2_full.reshape(raster_stack.shape[1], raster_stack.shape[2])

    # Save each component as a new GeoTIFF
    profile.update(dtype=rasterio.float32)

    with rasterio.open('PC1_map.tif', 'w', **profile) as dst:
        dst.write(pc1_map, 1)

    with rasterio.open('PC2_map.tif', 'w', **profile) as dst:
        dst.write(pc2_map, 1)

    print("\nSuccessfully saved PC1_map.tif and PC2_map.tif.")

except FileNotFoundError:
    print("Error: Make sure 'density_map.tif' and 'NO2_Concentration.tif' are in the correct folder.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import geopandas as gpd
import rasterio
from rasterio.warp import reproject, Resampling
import numpy as np
import glob
import os
from scipy.stats import gaussian_kde

print("--- Starting Daily Pixel-to-Pixel Correlation Analysis ---")

# --- Step 1: Set up paths and load the master NO2 raster ---

# Path to the folder containing your daily shapefiles
shapefile_folder = 'daily_shps/'
no2_raster_path = 'NO2_Concentration.tif'

# Find all daily shapefiles
shapefile_paths = sorted(glob.glob(os.path.join(shapefile_folder, '*.shp')))

# Load the NO2 raster once to use as a template
with rasterio.open(no2_raster_path) as no2_src:
    no2_data = no2_src.read(1)
    no2_profile = no2_src.profile # Get metadata for resampling

if not shapefile_paths:
    print(f"Error: No shapefiles found in '{shapefile_folder}'. Please check the folder.")
else:
    print(f"Found {len(shapefile_paths)} daily shapefiles to process.")

# --- Step 2: Loop through each day, create a density map, and correlate ---

# A list to store the results
daily_results = {}

for shp_path in shapefile_paths:
    day_name = os.path.basename(shp_path).split('.')[0]
    print(f"\n--- Processing: {day_name} ---")

    try:
        # a) Create Daily Density Raster (in memory)
        daily_gdf = gpd.read_file(shp_path)
        x = daily_gdf.geometry.x
        y = daily_gdf.geometry.y

        if len(x) < 2:
            print("Skipping day: Not enough data points to create a density map.")
            continue

        kde = gaussian_kde(np.vstack([x, y]))

        # We need to match the grid of the NO2 raster for the KDE
        min_lon, min_lat, max_lon, max_lat = no2_src.bounds
        width, height = no2_src.width, no2_src.height

        x_grid, y_grid = np.mgrid[min_lon:max_lon:complex(width), min_lat:max_lat:complex(height)]
        grid_positions = np.vstack([x_grid.ravel(), y_grid.ravel()])
        density_values = kde(grid_positions)
        daily_density_raster = density_values.reshape(x_grid.shape)
        daily_density_raster = np.flipud(daily_density_raster.T)

        # b) Correlate the daily density raster with the NO2 raster
        density_flat = daily_density_raster.flatten()
        no2_flat = no2_data.flatten()

        valid_data_mask = (density_flat > 0) & (no2_flat != no2_profile['nodata'])
        density_valid = density_flat[valid_data_mask]
        no2_valid = no2_flat[valid_data_mask]

        if len(density_valid) < 2:
            print("Skipping day: Not enough overlapping data to calculate correlation.")
            continue

        correlation_matrix = np.corrcoef(density_valid, no2_valid)
        correlation_coefficient = correlation_matrix[0, 1]

        daily_results[day_name] = correlation_coefficient
        print(f"Correlation for {day_name}: {correlation_coefficient:.4f}")

    except Exception as e:
        print(f"An error occurred processing {day_name}: {e}")

# --- Step 3: Display Final Results ---
print("\n--- Final Daily Correlation Results ---")
if daily_results:
    for day, corr in daily_results.items():
        print(f"{day}: {corr:.4f}")
else:
    print("No results were calculated.")
```
